```
data.fillna(data.mean(),inplace=True)
data
```

```
new_data = data.apply(abs)
new_data
# new_data.isnull().sum().sum()
```

```
new_data = pd.get_dummies(new_data)
```

```
x = new_data.iloc[:,:-1]
x
y = new_data[['Class']]
y 
x.dtypes
```

```
x.isnull().sum().sum()
x.isna().sum().sum()
x.isin(['-']).sum().sum()
```

```
from sklearn.feature_selection import  SelectKBest
from sklearn.feature_selection import  chi2
best_features = SelectKBest(score_func=chi2,k=10)
fit =  best_features.fit(x,y)
fit
```

```
data_score = pd.DataFrame(fit.scores_)
data_score
data_col = pd.DataFrame(x.columns)
data_col
```

```
fs = pd.concat([data_col,data_score],axis=1)
fs.columns = ['V1','V2']
print(fs.nlargest(1,'V2'))

```

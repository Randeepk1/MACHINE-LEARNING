```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```
data = pd.read_csv("/content/titanic_train.csv")
data
```

```
data = data.fillna({'age':data['age'].mean(),'fare':data['fare'].mean()})
data
```

```
x = data.iloc[:,[0,1,3,4,5,6,7,8]]
x
y= data['survived']
y
X = pd.get_dummies(x)
X
```

```
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=25,stratify = y,random_state=22)
```

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit_transform(x_train)
sc.transform(x_test)
```

```
y_train.value_counts(normalize=True)
```

```
y_test.value_counts(normalize=True)
```

```
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
model = dt.fit(x_train,y_train)
model
y_pred = model.predict(x_test)
y_pred
```

```
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)*100
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
lg = LogisticRegression(max_iter=2000)
model1 = lg.fit(x_train,y_train)
model1
y_pred1 = model1.predict(x_test)
y_pred1
accuracy_score(y_pred1,y_test)*100
```

```
from sklearn.model_selection import cross_val_score
dt = DecisionTreeClassifier(random_state=25)
cross_val_score(dt,x_train,y_train,cv=10,scoring='accuracy').mean()

```

```
from sklearn.linear_model import LogisticRegression
ld =LogisticRegression(max_iter=2000,random_state=100)
cross_val_score(ld,x_train,y_train,cv=10,scoring='accuracy').mean()
```
